```python
import pandas as pd
import matplotlib.pyplot as plt
from mplsoccer import VerticalPitch
```

```python
df = pd.read_csv('erling_haaland.csv')
df.head()
```

```python
df['X'] = df['X'] * 100
df['Y'] = df['Y'] * 100
```

```python
df.head()
```

```python
#calculate the stats we will need display
#Total shoits
# Total goals
# Total XG
# XG per shot 
# Average distance of shots

total_shots = df.shape[0]
total_goals = df[df['result'] == 'Goal'].shape[0]
total_xG = df['xG'].sum()
xG_per_shot = total_xG / total_shots
points_average_distance = df['X'].mean()
actual_average_distance = 120 - (df['X'] * 1.2).mean()
print(points_average_distance, actual_average_distance)
```

```python
background_color='#0C0D0E'
import matplotlib.font_manager as font_manager

font_path = 'avro/Avro-Regular.ttf'
font_props = font_manager.FontProperties(fname=font_path)
```

```python
pitch  = VerticalPitch(
    pitch_type='opta',
    half = True,
    pitch_color=background_color,
    pad_bottom=.5,
    line_color='white',
    linewidth=.75,
    axis= True,
    label= True
)
```

```python
fig, ax = plt.subplots(figsize=(10, 7))
#Draw the pitch 
pitch.draw(ax = ax)
plt.show()
```
